## 1. Import libraries

```python
import tensorflow as tf
tf.config.experimental.list_physical_devices(device_type='GPU')
```

```python
import numpy as np
import time
import pandas as pd
import tensorflow_datasets as tfds
import PIL
import PIL.Image
```

## 2. Download and/or import data

```python
import pathlib
download = False

if download:
    dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
    data_dir = tf.keras.utils.get_file(origin=dataset_url, 
                                   fname='flower_photos', 
                                   untar=True)
    data_dir = pathlib.Path(data_dir)
else:
    data_dir =  pathlib.Path('/home/data/.keras/datasets/flower_photos')
```

```python
roses = list(data_dir.glob('roses/*'))
PIL.Image.open(str(roses[0]))
```

### Import data to python, resize to 180x180 px and split data

```python
batch_size = 32
img_height = 299
img_width = 299
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.3,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.3,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
ds_size = len(val_ds)
test_size = int(ds_size/2)
test_ds = val_ds.take(test_size)
val_ds = val_ds.skip(test_size).take(ds_size-test_size)
```

```python
nclasses = len(train_ds.class_names)
print(train_ds.class_names)
```

### Adapting the pictures to the correct sizes

```python
def preprocess(image, label):
    # Taken from the Géron's book
    #resized_image = tf.image.resize(image, [229, 229]) # Optionally resize the image
    final_image = tf.keras.applications.inception_resnet_v2.preprocess_input(image)
    return final_image, label
```

```python
print(len(train_ds))
```

```python
train_ds = train_ds.map(preprocess).prefetch(1)
val_ds = val_ds.map(preprocess).prefetch(1)
test_ds = test_ds.map(preprocess).prefetch(1)
```

```python
print(len(train_ds))
```

## 3. Training
### 3.1 Import a known model and add replace layers at the end

```python
base_model = tf.keras.applications.inception_resnet_v2.InceptionResNetV2(
    weights="imagenet",
    include_top=False,
    pooling="max")
x = tf.keras.layers.Dense(nclasses*2, activation = "relu")(base_model.output)
output = tf.keras.layers.Dense(nclasses, activation = "softmax")(x)
model = tf.keras.Model(inputs = base_model.input, outputs = output)
```

```python
# The base model layers are set to not trainable during the first training phase
for layer in base_model.layers:
    layer.trainable = False
```

```python
# We use the traditional Stochastic Gradient Descent with a momentum optimizer and a power scheduling
optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.2,
    momentum=0.9,
    decay=0.01)
model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer=optimizer,
    metrics=["accuracy"])
```

```python
tb_cb = tf.keras.callbacks.TensorBoard("./Results_tensorboard/Ex_14_10/" + time.strftime("model_Transfer_InceptionResNetV2_%Y_%m_%d_%H_%M_%S"))
```

### 3.2 Train a some epochs with the final layers only

```python
model.fit(train_ds, epochs=10, validation_data=val_ds, callbacks=[tb_cb])
```

### 3.3 Train over the entire dataset

```python
# The base model layers are set to trainable during the second training phase
for layer in base_model.layers:
    layer.trainable = True
```

```python
# We use the traditional Stochastic Gradient Descent with a momentum optimizer and a power scheduling
optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.2,
    momentum=0.9,
    decay=0.001)
model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer=optimizer,
    metrics=["accuracy"])
```

```python
tb_cb = tf.keras.callbacks.TensorBoard("./Results_tensorboard/Ex_14_10/" + time.strftime("model_Transfer_InceptionResNetV2_part2_%Y_%m_%d_%H_%M_%S"))
model.fit(train_ds, epochs=10, validation_data=val_ds, callbacks=[tb_cb])
```
